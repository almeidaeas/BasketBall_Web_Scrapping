# BUSCO AS INFORMAÇÕES DOS JOGADORES

```python
headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}

Nome = []
Nascimento = []
Posicao = []


for index, row in df.iterrows():
    #Carrego a página do jogador correspondente        
    page = "https://www.basketball-reference.com" + row["Ref"]
    pageTree = requests.get(page, headers=headers)
    pageSoup = BeautifulSoup(pageTree.content, 'html.parser')

    print(page)
        
    #Carrego o objeto contendo os dados
    nome = pageSoup.find_all("h1", {"itemprop": "name"})    
    Nome.append(nome[0].text)
    
    nascimento = pageSoup.find_all("span", {"itemprop": "birthDate"})    
    Nascimento.append(nascimento[0].text.strip())

    posicao = pageSoup.find_all("p", limit=4)        
    
    #Corrigo o indice para pegar a posição do jogador
    if "lb" in (posicao[3].text.replace('Position:', '').replace('Shoots:', '').replace('Left', '').replace('Right', '').replace('▪', '').strip()):        
        Posicao.append(posicao[2].text.replace('Position:', '').replace('Shoots:', '').replace('Left', '').replace('Right', '').replace('▪', '').strip())
    elif "Born" in (posicao[3].text.replace('Position:', '').replace('Shoots:', '').replace('Left', '').replace('Right', '').replace('▪', '').strip()):
        Posicao.append(posicao[1].text.replace('Position:', '').replace('Shoots:', '').replace('Left', '').replace('Right', '').replace('▪', '').strip())
    elif "Team" in (posicao[3].text.replace('Position:', '').replace('Shoots:', '').replace('Left', '').replace('Right', '').replace('▪', '').strip()):
        Posicao.append(posicao[1].text.replace('Position:', '').replace('Shoots:', '').replace('Left', '').replace('Right', '').replace('▪', '').strip())
    else:
        Posicao.append(posicao[3].text.replace('Position:', '').replace('Shoots:', '').replace('Left', '').replace('Right', '').replace('▪', '').strip())

        
df["Nome"] = Nome
df["Nascimento"] = Nascimento
df["Posicao"] = Posicao


df
```

# BUSCO AS ESTATÍSTICAS AVANÇADAS DOS JOGADORES

```python
Jogos = []
Minutos = []
PER = []
VORP = []

for index, row in df.iterrows():
    #Carrego a página do jogador correspondente        
    page = "https://www.basketball-reference.com" + row["Ref"]
    pageTree = requests.get(page, headers=headers)
    pageSoup = BeautifulSoup(pageTree.content, 'html.parser')
    comments = pageSoup.find_all(string=lambda text: isinstance(text, Comment))
    print(page)

    tables = []
    for each in comments:
        if 'table' in each:
            try:
                tables.append(pd.read_html(each, attrs = {'id': 'advanced'})[0])
            except:
                continue
    
    df_adv = None
    df_filter_adv = None
                
    df_adv = tables[0]
    df_filter_adv = df_adv[df_adv['Season'] == '2018-19'] 
    
    
    if len(df_filter_adv["PER"]) == 0:
        PER.append(0)
        VORP.append(0)
        Jogos.append(0)
        Minutos.append(0)
    else:
        PER.append(float(sum(df_filter_adv["PER"])/len(df_filter_adv["PER"])))
        VORP.append(float(sum(df_filter_adv["VORP"])/len(df_filter_adv["VORP"])))
        Jogos.append(float(sum(df_filter_adv["G"])/len(df_filter_adv["G"])))
        Minutos.append(float(sum(df_filter_adv["MP"])/len(df_filter_adv["MP"])))

        
# df["Per"] = PER
# df["Vorp"] = VORP
df["Jogos"] = Jogos
df["MinutosJogados"] = Minutos

df                
```
